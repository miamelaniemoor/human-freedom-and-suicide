```python
#Standardizing column names
efw_data.rename(columns={'Year': 'year', 'Countries': 'countries'}, inplace=True)

#Filtering the most important columns
columns_to_keep = [
    'year',
    'ISO_Code',
    'countries',
    'SUMMARY INDEX',
    'Rank',
    'Quartile',
    'Size of Government',
    'Gender Adjustment',
    'Legal System & Property Rights',
    'Sound Money',
    'Freedom to trade internationally',
    'Regulation'
]
filtered_efw_data = efw_data[columns_to_keep]
filtered_efw_data.head()
```

### Data from Kaggle containing suicide metrics 

```python
suicide_data = pd.read_csv("master.csv")
```

```python
suicide_data.shape
```

```python
suicide_data.head()
```

```python
suicide_data.tail()
```

```python
suicide_data.loc[500]
```

```python
#Standardizing column name
suicide_data.rename(columns={'country': 'countries'}, inplace=True)

#Removing age groups from the data, summing up the data across age groups
suicide_data_aggregated = suicide_data.groupby(['countries', 'year', 'sex'], as_index=False).agg({
    'suicides_no': 'sum',
    'population': 'sum',
    'suicides/100k pop': 'mean',
})

#Pivoting the data and creating column for male and female, instead of sex column
suicide_pivot = suicide_data_aggregated.pivot(
    index=['countries', 'year'], 
    columns='sex', 
    values='suicides_no'
).reset_index()

suicide_pivot.rename(columns={'male': 'male_suicides', 'female': 'female_suicides'}, inplace=True)


#Merging aggregated data with other columns of the dataset
columns_to_keep = [col for col in suicide_data.columns if col not in ['age', 'sex', 'suicides_no']]
suicide_data_combined = pd.merge(
    suicide_data[columns_to_keep].drop_duplicates(subset=['countries', 'year']),
    suicide_pivot,
    on=['countries', 'year'],
    how='inner'
)

# Removing unnecessary columns
suicide_data_combined = suicide_data_combined.drop(columns=['generation', 'country-year'])

suicide_data_combined.head()
```

```python
print(suicide_data_combined['HDI for year'].isna().sum())
```

```python
suicide_data_combined = suicide_data_combined.drop(columns=['HDI for year'])
```

### Uniting the data into one dataframe

```python
merged_data = pd.merge(filtered_efw_data, hfi_data, on=['countries', 'year'], how='inner')
merged_data = pd.merge(merged_data, suicide_data_combined, on=['countries', 'year'], how='inner')

pd.set_option('display.max_columns', None)
merged_data.head()
```

```python
merged_data = merged_data.drop(columns=['ISO_code'])
merged_data.rename(columns={'SUMMARY INDEX': 'EFW_index'}, inplace=True)
```

### Visualizing

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group data by region and calculate mean suicide rates
region_suicides = merged_data.groupby('region')['suicides/100k pop'].mean().sort_values()

# Plot bar chart
plt.figure(figsize=(12, 8))
region_suicides.plot(kind='bar', color='orange')
plt.title('Average Suicide Rates by Region')
plt.xlabel('Region')
plt.ylabel('Suicides per 100k Population')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Bin Gender Adjustment into categories
merged_data['gender_adjustment_bins'] = pd.cut(
    merged_data['Gender Adjustment'], 
    bins=[0, 0.5, 0.75, 1.0], 
    labels=['Low', 'Medium', 'High']
)

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(merged_data['Gender Adjustment'], merged_data['suicides/100k pop'], alpha=0.6, color='purple')
plt.title('Suicide Rates vs. Gender Adjustment')
plt.xlabel('Gender Adjustment')
plt.ylabel('Suicides per 100k Population')
plt.grid(True)
plt.show()
```
