### Data from Kaggle containing suicide metrics 

```python
suicide_data = pd.read_csv("master.csv")
```

```python
suicide_data.shape
```

```python
suicide_data.head()
```

```python
suicide_data.tail(15)
```

```python
suicide_data.columns = suicide_data.columns.str.strip()
suicide_data.loc[500]
```

```python
#Standardizing column name
suicide_data.rename(columns={'country': 'countries'}, inplace=True)

#Removing age groups from the data, summing up the data across age groups
suicide_data_aggregated = suicide_data.groupby(['countries', 'year'], as_index=False).agg({
    'suicides_no': 'sum',
    'population': 'sum',  # Total population for all age groups and sexes
    'suicides/100k pop': 'mean',
    'gdp_for_year ($)': 'first',    
    'gdp_per_capita ($)': 'first'
})

suicides_by_sex = suicide_data.groupby(['countries', 'year', 'sex'], as_index=False).agg({
    'suicides_no': 'sum'
})

#Pivoting the data and creating column for male and female, instead of sex column
suicide_pivot = suicides_by_sex.pivot(
    index=['countries', 'year'],
    columns='sex',
    values='suicides_no'
).reset_index()

suicide_pivot.rename(columns={'male': 'male_suicides', 'female': 'female_suicides'}, inplace=True)

# Merging aggregated population and suicide pivot table
suicide_data_combined = pd.merge(
    suicide_data_aggregated,
    suicide_pivot,
    on=['countries', 'year'],
    how='inner'
)

suicide_data_combined.head()
```

### Uniting the data into one dataframe

```python
merged_data = pd.merge(filtered_efw_data, hfi_data, on=['countries', 'year'], how='inner')
merged_data = pd.merge(merged_data, suicide_data_combined, on=['countries', 'year'], how='inner')

merged_data = merged_data.drop(columns=['ISO_code'])
merged_data.rename(columns={'SUMMARY INDEX': 'EFW_index'}, inplace=True)

pd.set_option('display.max_columns', None)
merged_data
```

```python
#Removing countries which have been in a war or conflict
countries_to_remove = [
    'Libya', 'Ivory Coast', 'Syria', 'Sudan', 'Lebanon', 
    'Somalia', 'Iraq', 'Central African Republic', 
    'Israel', 'Afghanistan', 'Tunisia', 'Chad'
]

# Filter out rows with these countries
merged_data[~merged_data['countries'].isin(countries_to_remove)]
```

### Visualizing

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group data by region and calculate mean suicide rates
region_suicides = merged_data.groupby('region')['suicides/100k pop'].mean().sort_values()

# Plot bar chart
plt.figure(figsize=(12, 8))
region_suicides.plot(kind='bar', color='orange')
plt.title('Average Suicide Rates by Region')
plt.xlabel('Region')
plt.ylabel('Suicides per 100k Population')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Scatter for female suicides
plt.scatter(merged_data['Gender Adjustment'], merged_data['female_suicides'], alpha=0.6, color='red', label='Female Suicides')
# Scatter for male suicides
plt.scatter(merged_data['Gender Adjustment'], merged_data['male_suicides'], alpha=0.6, color='blue', label='Male Suicides')

# Add titles and labels
plt.title('Gender Adjustment vs. Suicide Rates')
plt.xlabel('Gender Adjustment')
plt.ylabel('Number of Suicides')
plt.legend()
plt.grid(True, which='both', linestyle='--')
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select columns related to economic indicators and suicide rates
correlation_columns = [
    'suicides/100k pop', 'EFW_index', 'Size of Government', 
    'Legal System & Property Rights', 'Sound Money', 
    'Freedom to trade internationally', 'Regulation', 'gdp_per_capita ($)'
]

# Calculate correlation matrix
correlation_matrix = merged_data[correlation_columns].corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between Suicide Rates and Economic Indicators')
plt.show()
```

```python
# Scatter plot for suicide rates vs. GDP per capita
plt.figure(figsize=(8, 6))
plt.scatter(merged_data['gdp_per_capita ($)'], merged_data['suicides/100k pop'], alpha=0.7, color='blue')
plt.title('GDP Per Capita vs. Suicide Rates')
plt.xlabel('GDP Per Capita ($)')
plt.ylabel('Suicides per 100k Population')
plt.grid(True)
plt.show()
```

```python
# Group data by countries and calculate mean male and female suicides
gender_suicides = merged_data.groupby('ISO_Code')[['male_suicides', 'female_suicides']].mean()

# Plot stacked bar chart with ISO codes
gender_suicides.plot(kind='bar', stacked=True, figsize=(16, 8), color=['blue', 'magenta'])

# Add titles and labels
plt.title('Male and Female Suicides by ISO Code (Average)')
plt.xlabel('ISO Code')
plt.ylabel('Number of Suicides')
plt.legend(['Male Suicides', 'Female Suicides'])
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()  # Ensures everything fits well in the figure
plt.show()
```

```python
# Group data by year and calculate mean suicide rates
yearly_suicides = merged_data.groupby('year')['suicides/100k pop'].mean()

# Plot line chart
plt.figure(figsize=(10, 6))
yearly_suicides.plot(kind='line', marker='o', color='red')
plt.title('Suicide Rates Over Time')
plt.xlabel('Year')
plt.ylabel('Suicides per 100k Population')
plt.grid(True)
plt.show()
```

```python
# Assume 'merged_data' is your DataFrame
# Calculate the proportion of missing values for each column
missing_proportions = merged_data.isnull().mean()

# Set thresholds for dropping columns
drop_threshold = 0.80  # Drop if >80% NaNs
review_threshold = 0.50  # Highlight for review if 50%-80% NaNs

# Identify columns to drop and review
columns_to_drop = missing_proportions[missing_proportions > drop_threshold].index.tolist()
columns_to_review = missing_proportions[(missing_proportions > review_threshold) & (missing_proportions <= drop_threshold)].index.tolist()

# Output recommendations
print(f"Columns to drop (>{drop_threshold*100}% NaNs): {columns_to_drop}")
print(f"Columns to review for potential imputation ({review_threshold*100}-{drop_threshold*100}% NaNs): {columns_to_review}")

# Drop columns with too many NaNs
cleaned_data = merged_data.drop(columns=columns_to_drop)

# Optional: save the cleaned data
# cleaned_data.to_csv('cleaned_data.csv', index=False)
```
