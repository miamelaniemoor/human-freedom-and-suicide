### Data from Kaggle containing suicide metrics 

```python
suicide_data = pd.read_csv("master.csv")
```

```python
suicide_data.shape
```

```python
suicide_data.head()
```

```python
suicide_data.tail(15)
```

```python
suicide_data.columns = suicide_data.columns.str.strip()
suicide_data.loc[500]
```

```python
#Standardizing column name
suicide_data.rename(columns={'country': 'countries'}, inplace=True)

#Removing age groups from the data, summing up the data across age groups
suicide_data_aggregated = suicide_data.groupby(['countries', 'year'], as_index=False).agg({
    'suicides_no': 'sum',
    'population': 'sum',  # Total population for all age groups and sexes
    'suicides/100k pop': 'mean',
    'gdp_for_year ($)': 'first',    
    'gdp_per_capita ($)': 'first'
})

suicides_by_sex = suicide_data.groupby(['countries', 'year', 'sex'], as_index=False).agg({
    'suicides_no': 'sum'
})

#Pivoting the data and creating column for male and female, instead of sex column
suicide_pivot = suicides_by_sex.pivot(
    index=['countries', 'year'],
    columns='sex',
    values='suicides_no'
).reset_index()

suicide_pivot.rename(columns={'male': 'male_suicides', 'female': 'female_suicides'}, inplace=True)

# Merging aggregated population and suicide pivot table
suicide_data_combined = pd.merge(
    suicide_data_aggregated,
    suicide_pivot,
    on=['countries', 'year'],
    how='inner'
)

suicide_data_combined.head()
```

### Uniting the data into one dataframe

```python
merged_data = pd.merge(filtered_efw_data, hfi_data, on=['countries', 'year'], how='inner')
merged_data = pd.merge(merged_data, suicide_data_combined, on=['countries', 'year'], how='inner')

merged_data = merged_data.drop(columns=['ISO_code'])
merged_data.rename(columns={'SUMMARY INDEX': 'EFW_index'}, inplace=True)

pd.set_option('display.max_columns', None)
merged_data
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.base import BaseEstimator, TransformerMixin

# Load the dataset (replace 'working_dataframe.xlsx' with your file path if necessary)
file_path = 'working_dataframe.xlsx'
merged_data = pd.read_excel(file_path)

# Identify numerical and categorical columns in `merged_data`
numerical_columns = merged_data.select_dtypes(include=['float64', 'int64']).columns
categorical_columns = merged_data.select_dtypes(include=['object']).columns

# Define preprocessing pipelines
numerical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),  # Fill NaNs with median
    ('scaler', StandardScaler())                   # Standardize numerical data
])

categorical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Fill NaNs with most frequent value
    ('onehot', OneHotEncoder(handle_unknown='ignore'))     # Encode categorical variables
])

# Combine preprocessing pipelines
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_pipeline, numerical_columns),
        ('cat', categorical_pipeline, categorical_columns)
    ]
)

# Define a custom transformer to convert sparse to dense
class ToDenseTransformer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        return X.toarray() if hasattr(X, "toarray") else X

# Define PCA to retain 95% of variance
pca = PCA(n_components=0.95)

# Full pipeline: preprocessing + dense conversion + PCA
full_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('to_dense', ToDenseTransformer()),  # Convert sparse matrix to dense
    ('pca', pca)                         # Apply PCA
])

# Fit and transform the data
processed_data = full_pipeline.fit_transform(merged_data)

# Convert the processed data into a DataFrame
pca_columns = [f'PC{i+1}' for i in range(processed_data.shape[1])]
processed_df = pd.DataFrame(processed_data, columns=pca_columns)

# Display the first few rows of the processed PCA DataFrame
print(processed_df.head())

# Save the processed DataFrame for future use
processed_df.to_csv('processed_data.csv', index=False)
print("Processed data saved as 'processed_data.csv'.")
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Determine the optimal number of clusters using the Elbow method
inertia = []  # Inertia values for each k
k_range = range(1, 11)  # Test k from 1 to 10

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(processed_df)
    inertia.append(kmeans.inertia_)

# Plot the Elbow curve
plt.figure(figsize=(8, 6))
plt.plot(k_range, inertia, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

# Based on the Elbow method, choose the optimal k (e.g., k=3 or k=4)
optimal_k = 3  # Replace with your chosen k value
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(processed_df)

# Add cluster labels to the PCA DataFrame
processed_df['Cluster'] = clusters
```
